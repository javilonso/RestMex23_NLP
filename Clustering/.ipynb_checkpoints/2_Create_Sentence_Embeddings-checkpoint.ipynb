```python
import numpy as np
from sklearn.cluster import KMeans
from sentence_transformers import SentenceTransformer

# Load a pre-trained SentenceTransformer model
model = SentenceTransformer('sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2', device='cuda')

# Load your 1000 texts into a list
texts = df['News'].tolist()

# Encode the texts into numerical vectors
text_embeddings = model.encode(texts, batch_size=2048)
```

```python
with open('drive/My Drive/RestMex23/Clustering/emb_l12v2_spacy.npy', 'wb') as f:
    np.save(f, text_embeddings)
```
