```python
file_path='train/Rest_Mex_Sentiment_Analysis_2023_Train.xlsx'

import pandas as pd

# Replace 'example.xlsx' with the name of your Excel file
df = pd.read_excel(file_path, engine='openpyxl')

# Print the first five rows of the DataFrame
print(df.head())
```

```python
datapd = df.copy()
print(f'Polarity values({datapd.Polarity.count()}):\n{datapd.Polarity.value_counts()}\n')
print(f'Atraction values({datapd.Type.count()}):\n{datapd.Type.value_counts()}\n')
print(f'Atraction values({datapd.Country.count()}):\n{datapd.Country.value_counts()}\n')
print(f'Title values:\n{datapd.Title.str.len().describe()}\n')
print(f'Opinion values:\n{datapd.Review.str.len().describe()}')
```

```python
datapd
```

```python
# filter non empty title
datapd = datapd[datapd.Title.str.len() > 0]
# filter non empty opinion
datapd = datapd[datapd.Review.str.len() > 0]
```

```python
datapd = datapd[datapd.Review.str.len() < 5000]
```

```python
datapd
```

```python
for i in range(5):
    datapd.loc[datapd.Polarity == i+1, 'Polarity'] = i
    
datapd.loc[datapd.Type == "Hotel", 'Type'] = 0
datapd.loc[datapd.Type == "Restaurant", 'Type'] = 1
datapd.loc[datapd.Type == "Attractive", 'Type'] = 2

datapd.loc[datapd.Country == "Mexico", 'Country'] = 0
datapd.loc[datapd.Country == "Cuba", 'Country'] = 1
datapd.loc[datapd.Country == "Colombia", 'Country'] = 2

datapd["Title_Review"] = datapd["Title"] + " " + datapd["Review"]
```

```python
import pandas as pd
import numpy as np

def oversample_df(df, label_column, max_samples=None):
    # Count the number of examples per label
    counts = df[label_column].value_counts()

    # Compute the maximum number of examples to add per label
    if max_samples is None:
        max_samples = counts.max()

    # Initialize a list to store the oversampled dataframes
    df_list = []

    # Iterate over the labels
    for label in counts.index:
        # Get the number of examples for the current label
        label_count = counts[label]

        # Compute the number of examples to add
        num_to_add = min(max_samples - label_count, label_count)

        if num_to_add > 0:
            # Get the examples for the current label
            label_df = df[df[label_column] == label]

            # Choose the examples to add randomly
            idx = np.random.choice(label_df.index, size=num_to_add, replace=True)
            examples_to_add = label_df.loc[idx]

            # Append the original examples and the new examples to the list
            df_list.append(label_df)
            df_list.append(examples_to_add)

    # Concatenate the dataframes in the list to create the oversampled dataframe
    oversampled_df = pd.concat(df_list, axis=0)

    # Shuffle the dataframe to ensure that the examples are in a random order
    oversampled_df = oversampled_df.sample(frac=1, random_state=42)

    return oversampled_df
```

```python
max_samples = 50000  # or any other desired number of samples per class
oversampled_df = oversample_df(datapd, 'Polarity', max_samples=max_samples)
```

```python
frames = [datapd, oversampled_df]
  
result = pd.concat(frames)
```

```python
df_resampled = result
```

```python
print(f'Polarity values({df_resampled.Polarity.count()}):\n{df_resampled.Polarity.value_counts()}\n')
print(f'Atraction values({df_resampled.Type.count()}):\n{df_resampled.Type.value_counts()}\n')
print(f'Atraction values({df_resampled.Country.count()}):\n{df_resampled.Country.value_counts()}\n')
print(f'Title values:\n{df_resampled.Title.str.len().describe()}\n')
print(f'Opinion values:\n{df_resampled.Review.str.len().describe()}')
```
