```python
# Download dataset
from datasets import Dataset, load_dataset
raw_dataset = load_dataset("javilonso/restmex23-test")
```

```python
from transformers import pipeline
pred_pipeline = pipeline("text-classification", model="javilonso/rm23_ctry1", device='cuda:0', tokenizer = "cardiffnlp/twitter-xlm-roberta-base")
```

```python
# Instances to perform inference and submit to competition
raw_dataset['train']
```

```python
pred_pipeline_output = pred_pipeline(raw_dataset['train']["Title_Review"], truncation=True, max_length=512)
```

```python
# label_mapping = {'LABEL_2': "Attractive", 'LABEL_1': "Restaurant", 'LABEL_0': "Hotel"}
label_mapping = {'LABEL_2': "Colombia", 'LABEL_1': "Cuba", 'LABEL_0': "Mexico"}
# label_mapping = {'LABEL_4': 5, 'LABEL_3': 4,'LABEL_2': 3, 'LABEL_1': 2, 'LABEL_0': 1}

# Perform inference and match model labels with actual competition labels
for item in pred_pipeline_output:
    item['label'] = label_mapping.get(item['label'], item['label'])
```

```python
predictions = [item['label'] for item in pred_pipeline_output]
```

```python
import numpy as np
np.save('pred_pol4_epoch7.npy', np.array(predictions), allow_pickle=True)
```
